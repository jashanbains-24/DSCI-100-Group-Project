<font size = "4"> **1. Data Description** <font>

The total dataset consists of two different files;
- "players.csv" describes the individual characteristics of the participating players, such as their age, first name, gender, experience within the game, their "hashedEmail", whether they are subscribed, and how many hours they have played.
- "sessions.csv" tracks gameplay statistics, specifically, start and end time for each player. Also includes hashedEmail.




Players Dataset (players.csv) (196 rows x 7 columns)
- `experience` (chr) – Player level (Pro, Veteran, Regular, Amateur)
- `subscribe` (lgl) – Subscribed to newsletter? (T/F)
- `hashedEmail` (chr) – Unique player identifier
- `played_hours` (dbl) – Total hours played
- `name` (chr) – First name
- `gender` (chr) – Player’s gender
- `Age` (dbl) – Player’s age (yrs) (2 missing values)

Sessions Dataset (sessions.csv) (1535 rows x 5 columns)
- `hashedEmail` (chr) – Same as above
- `start_time` (chr) – Session start time (to be converted)
- `end_time` (chr) – Session end time (to be converted, 2 missing values)
- `original_start_time` (dbl) – Unix timestamp for start_time
- `original_end_time` (dbl) – Unix timestamp for end_time (2 missing values)

Key Issues & Approach:
Some values are missing (Age, end_time), and playtime data is skewed by inactive players. Merging datasets via hashedEmail enables timestamp cleanup, missing value imputation, and insights like session duration and peak hours to analyze engagement by player type.

<font size = "4"> **2. Question(s)** <font>

The broad question we want to address is: Which types of players contribute the most gameplay data, and how can we target them for recruitment?

Our specific question is: Do certain characteristics of player data, such as playtime or age, determine whether players could belong to a specific skill level, gender, or even predict whether they are subscribed to the newsletter?

To answer this, we will analyze total hours played, session durations, `subscription`s, `Age`s, etc. to determine if distinct characteristis separate player `experience` levels. Merging players.csv and sessions.csv via `hashedEmail` will allow for cleaner timestamps, calculating average session length, limit duplicates, and handling missing values. This will help define clear differences between each category and enable future predictions as new or returning players enter the game.

<font size = "4"> **3.  Exploratory Data Analysis and Visualization** <font> 

```r
# data extraction and manipulation

# loading packages
library(tidyverse)
library(readr)
library(ggplot2)
library(dplyr)
library(tidyr)
library(lubridate)
library(RColorBrewer)

# reading and calling the data
players <- read_csv("https://raw.githubusercontent.com/jashanbains-24/DSCI-100-Group-Project/refs/heads/main/players.csv")
sessions <- read_csv("https://raw.githubusercontent.com/jashanbains-24/DSCI-100-Group-Project/refs/heads/main/sessions.csv")

# merging at hashedEmail, grouped by hashedEmail for each player
full_set <- left_join(players, sessions)

# separating times into dates and times, as well as getting rid of original_start_time and original_end_time
full_set_improved <- full_set |>
            separate(start_time, into = c("start_date", "start_time"), sep = " ") |>
            separate(end_time, into = c("end_date", "data_time"), sep = " ") |>
            select(-original_start_time, -original_end_time)
full_set_improved

# computing means for all quantitatives
mean_ages <- full_set_improved |>
                    select(Age) |>
                    summarize(average_age = mean(Age, na.rm = TRUE))
mean_playtime <- full_set_improved |>
                    select(played_hours) |>
                    summarize(average_playtime = mean(played_hours, na.rm = TRUE))

# computing mins and maxes for all quantitatives
max_age <- full_set_improved |>
            select(Age) |>
            summarize(max_age = max(Age, na.rm = TRUE))
min_age <- full_set_improved |>
            select(Age) |>
            summarize(min_age = min(Age, na.rm = TRUE))

max_playtime <- full_set_improved |>
            select(played_hours) |>
            summarize(max_playtime = max(played_hours, na.rm = TRUE))
min_playtime <- full_set_improved |>
            select(played_hours) |>
            summarize(min_playtime = min(played_hours, na.rm = TRUE))

# computing standard deviation for all quantitatives
sd_age <- full_set_improved |>
                    select(Age) |>
                    summarize(sd_age = sd(Age, na.rm = TRUE))
sd_playtime <- full_set_improved |>
                    select(played_hours) |>
                    summarize(sd_playtime = sd(played_hours, na.rm = TRUE))

# visualizations
options(repr.plot.width = 9, repr.plot.height = 7)

player_ages <- ggplot(players, aes(x = Age, color = Age)) +
                  geom_histogram(binwidth = 5, fill = "steelblue", color = "black") +
                  labs(title = "Age Distribution of Players", x = "Age", y = "Count") + 
                    theme(text = element_text(size = 20))
player_ages


hrs_v_age <- ggplot(full_set_improved, aes(x = Age, y = played_hours)) +
                geom_point(alpha = 0.4) +
                labs(x = "Player Age (yrs)", y = "Playtime (hrs)", title = "Hours Played vs. Player Age") +
                theme(text = element_text(size = 20))
hrs_v_age


hours_for_type <- full_set_improved |>
                    group_by(experience) |>
                    summarize(average_playtime = mean(played_hours, na.rm = TRUE)) |>
                    ggplot(aes(x = experience, y = average_playtime, fill = experience)) +
                    geom_bar(stat = "identity") +
                    labs(x = "Player Experience Level", y = "Average Playtime (hrs)", title = "Average Playtime for Each Player Type") +
                    theme(text = element_text(size = 20)) + 
                    scale_fill_brewer(palette = "PRGn")
hours_for_type


avg_playtime_gender <- players |>
                        group_by(gender) |>
                        summarize(avg_playtime = mean(played_hours, na.rm = TRUE))

gender_and_hours <- avg_playtime_gender |>
                        ggplot(aes(x = gender, y = avg_playtime, fill = gender)) +
                        geom_bar(stat = "identity") +
                        labs(x = "Player Gender", y = "Average Playtime (hrs)", title = "Average Hours Played for Each Gender") +
                        theme(text = element_text(size = 20)) +
                        theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1)) +
                        scale_fill_brewer(palette = "Dark2")
gender_and_hours
```

<font size = "4"> **4. Overall Plan and Approach** <font> 

We will use KNN classification to predict player categories based on individual characteristics. 

$Why$ $KNN?$
- No Assumptions – Works without predefined data distribution.
- Captures Non-Linearity – Identifies complex patterns in playtime and experience.
- Interpretable – Groups similar players naturally.

$Limitations$
- Requires Scaling of Data – Distance-based method needs normalization.
- Computationally Intensive – Performance slows with large datasets, worsening classification effectiveness over time.
- Sensitive to Imbalanced Data – May struggle with underrepresented player types.

$Process$ $and$ $Validation$
- 5-fold cross-validation will tune KNN and assess accuracy, recall, and precision.
- RMSE & RMSPE (via KNN or linear regression) will verify classification effectiveness.
- Data split proportion of anywhere from 0.75 to 0.6 will optimize performance.
- Standard workflow (recipe, training, evaluation, and visualization) ensures model robustness.
